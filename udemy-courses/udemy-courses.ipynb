# 1. What are all different subjects for which Udemy is offering courses?

# 2. Which subject has the maximum number of courses?

# 4. Show all the courses which are Paid

# 5. Which are `Top Selling` Courses?

```python
# we find the top selling courses by using the num_subscribers column
course.sort_values(by='num_subscribers', ascending=False).head(10)
```

```python
# find the names of the top 10 courses
top_courses = list(course.sort_values(by='num_subscribers', ascending=False)['course_title'])[:11]
top_courses
```

# 6. Which are `Least Selling` Courses ?

```python
# we find the least selling courses by using the num_subscribers column
course.sort_values(by='num_subscribers', ascending=True).head(10)
```

```python
# as there are a lot of courses with 0 subscribers, we will find total count of courses with 0 subscribers
zero_courses_count = course[course['is_paid'] == False]['is_paid'].count()
zero_courses_count
```

```python
# display 15 courses that has zero subscribers
zero_courses = list(course.sort_values(by='num_subscribers', ascending=True)['course_title'])[:16]
zero_courses
```

# 10. What is the MAXIMUM number of subscribers for each Level of courses?

```python
level_subs_series = course.groupby('level')['num_subscribers'].max()
level_subs_df = level_subs_series.reset_index().rename(columns={'num_subscribers':'max_num_subscribers'})
level_subs_df
```

```python
sorted_level_subs = level_subs_df.sort_values(by='max_num_subscribers', ascending=False)
sorted_level_subs
```
